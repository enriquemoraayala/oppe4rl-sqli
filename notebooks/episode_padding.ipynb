```python
#reading all the files

#reading the first file
df = pd.read_csv('../src/outputs/run_history_19.csv', sep=';', index_col=0)
df_emb = pd.read_csv('../src/outputs/run_history_emb_idx_19.csv', sep=';', index_col=0)
np_emb_state = np.load('../src/outputs/emb_states_19.npy', allow_pickle=True)
np_emb_next_state = np.load('../src/outputs/next_states_19.npy', allow_pickle=True)
```

```python
n_win = df[df['reward'] >= 0].episode.nunique()
episodes = list(df.episode.unique())
print(len(episodes))
print(n_win)
```

```python
columns = []
for i in range(30):
    columns.append('s_' + str(i))
    columns.append('a_' + str(i))
    columns.append('r_' + str(i))
print(columns)
```

```python
df_episodes_padded = pd.DataFrame(columns=columns)
```

```python
df_episodes_padded.head()
```

```python
neutral_action = 99
for ep in episodes:
    row = []
    df_ = df[df['episode'] == ep].copy()
    
    for i,step in df_.iterrows():
        state = np_emb_state[i]
        row.append(state)
        row.append(step['action'])
        row.append(step['reward'])
            
    df_size = len(df_)
    if df_size < 29:
        for i in range(df_size, 30):
            row.append(state)
            row.append(neutral_action)
            row.append(0.0)
    print(len(row))
    df_episodes_padded = df_episodes_padded.append(pd.Series(row, index =columns), ignore_index=True)
  
    break
  
```
